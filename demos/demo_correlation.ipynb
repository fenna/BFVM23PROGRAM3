```python
import pandas as pd
import numpy as np
from datetime import datetime
import neurokit2 as nk
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
ecg12 = nk.ecg_simulate(duration=10, method="multileads")
nk.signal_plot(ecg12, subplots=True)
```

```python
ecg12
```

```python
ecg12.boxplot()
```

```python
sns.set(style="whitegrid")  # Set the style if needed
corr_ecg = ecg12.corr(method='spearman')

# Customize the heatmap of the corr_meat correlation matrix and rotate the x-axis labels
fig = sns.clustermap(corr_ecg,
                     row_cluster=True,
                     col_cluster=True,
                     figsize=(6, 6))

plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(), rotation=90)
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(), rotation=0)
plt.savefig('clustermap.png')
plt.show()
```

```python
# Customize the heatmap of the corr_meat correlation matrix and rotate the x-axis labels
fig = sns.heatmap(corr_ecg)
plt.savefig('heatmapmap.png')
plt.show()
```
